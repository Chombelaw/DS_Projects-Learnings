```python
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], bins = 50, kde=False)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x= tips['total_bill'], y = tips['tip'], kind='hex')
```

```python
sns.jointplot(x = tips['total_bill'], y = tips['tip'], kind = 'kde')
```

```python
sns.jointplot(x = tips['total_bill'], y= tips['tip'], kind = 'reg')
```

```python
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (10,10)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker',diag_kind='kde')
```

```python
sns.rugplot(tips['total_bill'],color = 'r')
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
sns.distplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
