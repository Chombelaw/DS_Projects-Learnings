```python
X = df[['ZN','CRIM', 'RM', 'AGE','LSTAT','DIS']]
```

```python
y = df['price']
```

```python
X.head(),y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
lreg.fit(X_train, y_train)
```

```python
yIntercept= lreg.intercept_
```

```python
coeff = lreg.coef_
```

```python
yIntercept
```

```python
coeff
```

```python
X.columns
```

```python
coeffDf = pd.DataFrame(data = coeff, columns=['Cofeecient'],index =X.columns )
```

```python
coeffDf
```

```python
ypred = lreg.predict(X_test)
```

```python
plt.scatter(y_test,ypred)
```

```python
sns.distplot(y_test-ypred)
```

```python
from sklearn import metrics
```

```python
print('MAE',metrics.mean_absolute_error(y_test,ypred))
print('MSE',metrics.mean_squared_error(y_test,ypred))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,ypred)))
```
