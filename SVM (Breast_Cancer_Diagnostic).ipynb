```python
df = df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'diagnosis']]
```

```python
df.head()
```

```python
sns.pairplot(data =df, hue='diagnosis')
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('diagnosis',axis = 1)
y= df['diagnosis']
```

```python
X.head()
```

```python
y.head()
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svm_m = SVC(C = 1000,gamma=0.0001)
```

```python
svm_m.fit(X_train,y_train)
```

```python
y_pred = svm_m.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
param = {'C':[1,10,100,1000],
         'gamma':[0.1,0.01,0.001,0.0001]}
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
grid = GridSearchCV(estimator = SVC(), param_grid = param,  verbose=2)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
gred_yPred = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,gred_yPred))
print(classification_report(y_test,gred_yPred))
```
