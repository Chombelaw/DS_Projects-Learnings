```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot( x = tips['day'], y=tips['total_bill'],jitter=True, palette='coolwarm', hue= tips['sex'])
```

```python
sns.swarmplot(x='day', y= 'total_bill', data = tips,hue='sex')
```

```python
sns.swarmplot(x='total_bill', y= 'day', data = tips,hue='time', orient=True)
```

```python
sns.swarmplot(x='total_bill', y= 'day', data = tips,hue='time')
```

```python
sns.boxplot(x='day', y= 'total_bill', data = tips,hue='sex',palette='rainbow')
```

```python
sns.violinplot(x='day', y= 'total_bill', data = tips,hue='sex',palette='rainbow')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex',palette='rainbow',split=True)
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex',palette='rainbow')
```

```python
import numpy as np
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, estimator=np.std, hue = 'sex',palette='rainbow')
```

```python
sns.countplot(x = 'day', data=tips,hue='sex' )
```
